```python
import pandas as pd
```

```python
import numpy as np

# Create dummy data
data = {
    'sex': np.concatenate([np.ones(42), np.zeros(58)]).astype(int), # Modified to have 42 males and 58 females
    'age': np.random.randint(18, 80, 100),
    'obstruct': np.random.randint(0, 2, 100),
    'outcome': np.random.randint(0, 2, 100),
    'TRTMT': np.random.randint(0, 2, 100)
}

# Shuffle the sex column to mix male and female patients
np.random.shuffle(data['sex'])


# Create a pandas DataFrame
dummy_df = pd.DataFrame(data)

# Display the first few rows of the dummy DataFrame
print("Dummy DataFrame:")
display(dummy_df.head())

# Display the data types of the columns
print("\nData types:")
display(dummy_df.dtypes)
```

```python
# Save the dummy DataFrame to a CSV file
dummy_df.to_csv('dummy_data.csv', index=False)

print("dummy_data.csv created successfully.")
```
